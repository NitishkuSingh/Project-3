# 🔍 Exploratory Data Analysis (EDA)

# 🟧 Box Plots: Show the distribution of data based on five summary statistics: minimum, first quartile (Q1), median, third quartile (Q3), and maximum.
1] Minimum: The smallest value in the dataset, excluding outliers. It's shown as the leftmost end of the "whisker" on the left side of the box.

2] First Quartile (Q1): This is the 25th percentile, meaning 25% of the data falls below this value. It’s represented by the left edge of the box.

3] Median: Also known as the 50th percentile, it’s the middle value of the dataset, dividing it into two equal halves. The median is shown as a line within the box.

4] Third Quartile (Q3): This is the 75th percentile, meaning 75% of the data falls below this value. It’s represented by the right edge of the box.

5] Maximum: The largest value in the dataset, excluding outliers. It’s shown as the rightmost end of the "whisker" on the right side of the box.

#Outliers: These are data points that fall outside of the "whiskers" and are often plotted as individual points.

#Interquartile Range (IQR): The range between Q1 and Q3, representing the middle 50% of the data. It’s the height of the box.

```python
plt.figure(figsize=(8, 6), dpi=100)

df_remove = df.drop('Rank', axis=1)

sns.boxplot(data=df_remove, orient='h', palette="coolwarm")

plt.title('Distribution of Medals', fontsize=16, fontweight='bold')
plt.ylabel('Medal Type', fontsize=14)
plt.xlabel('Number of Medals', fontsize=14)

plt.show()
```

# 🟧 Histograms: Visual representation of the distribution of numerical data.

```python
plt.figure(figsize=(14, 10), dpi=120)

columns = ["Gold", "Silver", "Bronze", "Total"]

for i, col in enumerate(columns):
    plt.subplot(2, 2, i+1)  
    sns.histplot(data=df, x=col, bins=50, kde=True, color=sns.color_palette("Set2")[i])
    plt.title(f"Distribution of {col} Medals", fontsize=14, fontweight='bold')
    plt.xlabel(f"Number of {col} Medals", fontsize=12)
    plt.ylabel("Frequency", fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()

plt.show()
```

# 📊 Bar plot: Is used to compare the values of different categories visually, with the length of each bar representing the magnitude of each category.

```python
top_gold_countries = df.sort_values(by='Gold', ascending=False)

top_gold_countries = top_gold_countries[['Country', 'Gold']].head(10)
```

```python
plt.figure(figsize=(8,6), dpi=100)
sns.barplot(data=top_gold_countries, x='Country', y='Gold', color='Gold')

plt.title("Gold Medals by Top 10 Countries", fontsize=16, fontweight='bold')
plt.xlabel("Country", fontsize=18)
plt.ylabel("Gold Medals", fontsize=14)
plt.xticks(rotation=90)
```

```python
top_silver_countries = df.sort_values(by="Silver", ascending=False)
top_silver_countries = top_silver_countries[['Country', 'Silver']].head(10)

plt.figure(figsize=(8,6), dpi=100)
sns.barplot(data=top_silver_countries, x='Country', y='Silver', color='Silver')

plt.title("Silver Medals by Top 10 Countries", fontsize=16, fontweight='bold')
plt.xlabel("Country", fontsize=14)
plt.ylabel("Silver Medals", fontsize=14)
plt.xticks(rotation=90)
```

```python
top_bronze_countries = df.sort_values(by='Bronze', ascending=False)
top_bronze_countries = top_bronze_countries[['Country', 'Bronze']].head(10)

plt.figure(figsize=(8,6), dpi=100)
sns.barplot(data=top_bronze_countries, x='Country', y='Bronze', color="#CD7F32")

plt.title("Bronze Medals by Top 10 Countries", fontsize=16, fontweight='bold')
plt.xlabel("Country", fontsize=14)
plt.ylabel("Bronze Medals", fontsize=14)
plt.xticks(rotation=90)
```

```python
top_total_countries = df.sort_values(by='Total', ascending=False)
top_total_countries = top_total_countries[['Country', 'Total']].head(10)

plt.figure(figsize=(8,6), dpi=100)
sns.barplot(data=top_total_countries, x='Country', y='Total')

plt.title("Total Medals by Top 10 Countries", fontsize=16, fontweight='bold')
plt.xlabel("Country", fontsize=14)
plt.ylabel("Total Medals", fontsize=14)
plt.xticks(rotation=90)
```

# 🟧 Pairplot: visualizes relationships between pairs of variables in a dataset using scatter plots and histograms, helping to identify correlations and distributions across multiple dimensions.

```python
plt.figure(figsize=(10, 8), dpi=100)

pairplot = sns.pairplot(data=df_remove,
                        plot_kws={'alpha': 0.7, 's': 70},  
                        diag_kind='kde',                  
                        diag_kws={'shade': True, 'linewidth': 2},  
                        height=2.5)                        

plt.suptitle("Distribution of Medals Across Corpus", fontsize=22, fontweight='bold', color='darkblue')

plt.subplots_adjust(top=0.95)

sns.set(style="whitegrid", rc={"grid.linestyle": "--"})

plt.show()
```

# 🥧 Pie chart: shows the proportions of different categories as slices of a circle, with each slice representing a category's contribution to the whole.

```python
df2 = df[:10].copy()
new_row = pd.DataFrame(data={
    'Country': ['others'],
    'Total': [df['Total'][10:].sum()]
})
df2 = pd.concat([df2, new_row])

plt.figure(figsize=(8, 6), dpi=100)

plt.pie(df2['Total'], labels=df2['Country'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette("Set2", len(df2)))

plt.title('Distribution of Total Medals by Country')

plt.show()
```
